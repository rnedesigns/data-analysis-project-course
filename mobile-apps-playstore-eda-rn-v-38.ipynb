**NOTES** - The column `Android version` has been added and the `Android Ver` has been removed from the `playstore_updated_df` dataframe. I have converted the object data type to the `int` type for the `Android version`. For ex: 403 means the `Android version` is 4.0.3

```python
# To convert Android Version to Float, first remove the characters from the text:
android_ver_to_numeric = playstore_df['Android Ver'].astype(str).str.replace(' and up', '').str.replace('.', '')
android_ver_to_numeric.astype(str).str.replace('Varies with device', '0')
```

```python
# Converting 'Android version' numeric type of `float`:
android_ver_to_numeric = pd.to_numeric(android_ver_to_numeric, errors='coerce')
```

```python
# To check the data type 'Android version' column: 
android_ver_to_numeric.dtype
```

```python
# Creating new column 'Android version' and assigning formatted data of float type:
playstore_updated_df['Android version'] = android_ver_to_numeric
```

```python
# Convert float numeric type to 'int':
playstore_updated_df['Android version'] = playstore_updated_df['Android version'].fillna(0.0).astype(int)
playstore_updated_df['Android version']
```

```python
# To delete the 'Android Ver' from `playstore_updated_df` dataframe:
playstore_updated_df.drop(columns='Android Ver', inplace=True)
```

```python
# To view the modified dataframe `playstore_updated_df`:
playstore_updated_df
```

```python
# To confirm the column 'Android version' been changed to `int` type:
playstore_updated_df['Android version'].dtype
```
